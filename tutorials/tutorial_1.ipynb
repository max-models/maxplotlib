```python
import maxplotlib.canvas.canvas as canvas

%load_ext autoreload
%autoreload 2
```

```python
c = canvas.Canvas(width=800, ratio=0.5)
sp = c.add_subplot(grid=True, xlabel='(x - 10) * 0.1', ylabel='10y',yscale=10,xshift=-10,xscale=0.1)
sp.add_line([0, 1, 2, 3], [0, 1, 4, 9], label="Line 1")
sp.add_line([0, 1, 2, 3], [0, 2, 3, 4], linestyle='dashed', color='red', label="Line 2")
c.plot()
c.savefig(filename='plot_00.pdf')
```

```python
c = canvas.Canvas(ncols=2,nrows=2, width=3000, ratio=0.5)
sp = c.add_subplot(grid=True)
c.add_subplot(row=1)
sp2 = c.add_subplot(row=1,legend=False)
sp.add_line([0, 1, 2, 3], [0, 1, 4, 9], label="Line 1")
sp2.add_line([0, 1, 2, 3], [0, 2, 3, 4], linestyle='dashed', color='red', label="Line 2")
c.plot(backend='matplotlib')
c.plot(backend='plotly')
c.savefig(filename='plot_01.pdf')
```
